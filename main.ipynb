### Fixing Outliers

```python
# study hours cannot be negative 
df.loc[df['Study Hours per Week'] < 0, 'Study Hours per Week'] = np.nan

# attendance between 0-100
df.loc[(df['Attendance Rate'] < 0) | (df['Attendance Rate'] > 100), 'Attendance Rate'] = np.nan

# pervious grades (assuming 0-100 scale)
df.loc[df['Previous Grades'] > 100,' Previous Grades'] = np.nan
```

```python
# impute numerical column with median
num_col = ['Study Hours per Week', 'Attendance Rate', 'Previous Grades']
for col in num_col:
      df[col] = df[col].fillna(df[col].median())
```

```python
# impute categorical column with mode
cat_col = ['Participation in Extracurricular Activities', 'Parent Education Level', 'Passed']
for col in cat_col:
      df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df = df.drop_duplicates()
```
